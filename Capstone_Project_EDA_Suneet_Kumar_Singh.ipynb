### Variables Description

ID – Unique identifier for each logistics record.

Project Code – Code representing the project or initiative tied to the shipment.

PO / SO # – Purchase Order or Sales Order associated with the shipment.

ASN/DN # – Advanced Shipment Notice or Delivery Note reference.

Country – Destination country for the shipment (e.g., Côte d’Ivoire, Vietnam).

Managed By – Specifies which team manages the project (e.g., PMO – US).

Fulfill Via – Method of fulfillment, such as direct drop shipment.

Vendor INCO Term – Incoterms defining the shipment agreement (e.g., EXW, FCA).

Shipment Mode – The mode of transportation used (e.g., air).

PQ First Sent to Client Date – The date the Price Quotation was first sent to the client (some records indicate “Pre-PQ Process”).

PO Sent to Vendor Date – Date the purchase order was sent to the vendor.

Scheduled Delivery Date – The planned date for delivery.

Delivered to Client Date – The actual date the shipment was delivered.

Delivery Recorded Date – Date when delivery was recorded.

Product Group – General classification of the product (e.g., HIV, ARV).

Sub Classification – Further categorization of the product (e.g., HIV test, Pediatrics).

Vendor – The company supplying the goods (e.g., RANBAXY Fine Chemicals LTD, Aurobindo Pharma Limited).

Item Description – Detailed description of the item being shipped.

Molecule/Test Type – The type of test or pharmaceutical molecule (e.g., HIV, Nevirapine).

Brand – The brand name of the product (e.g., Reveal, Generic).

Dosage – The dosage strength of the pharmaceutical product (if applicable).

Dosage Form – The form in which the product is delivered (e.g., test kit, oral suspension).

Unit of Measure (Per Pack) – The number of units per pack.

Line Item Quantity – The total number of units shipped.

Line Item Value – The total value of the line item in the shipment.

Pack Price – Price per pack of the item.

Unit Price – Price per individual unit of the item.

Manufacturing Site – The location where the item was manufactured (e.g., Ranbaxy Fine Chemicals LTD, Aurobindo Unit III, India).

First Line Designation – Indicates whether the product is designated for first-line use.

Weight (Kilograms) – The weight of the shipment.

Freight Cost (USD) – The cost of shipping the item in U.S. dollars.

Line Item Insurance (USD) – Insurance cost for the line item.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
data_copy=data.copy()
````

````python
# converting column names to lower case, removing white spaces, replacing spaces with underscore and removing special characters
data_copy.columns = (
    data_copy.columns.str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace(r"[^\w_]", "", regex=True)
)
````

````python
data_copy.columns
````

````python
#filling missing values in line item insurance with median values
data_copy["line_item_insurance_usd"].fillna(data_copy["line_item_insurance_usd"].median(), inplace=True)

#filling missing values in shipment mode with mode
data_copy["shipment_mode"].fillna(data_copy["shipment_mode"].mode()[0], inplace=True)

#filling missing values in dosage with Not Available
data_copy["dosage"].fillna("Not Available", inplace=True)

data_copy.isnull().sum()
````

````python
#converting dates from string to datetime format
date_columns = [col for col in data_copy.columns if "date" in col]

for col in date_columns:
    data_copy[col] = pd.to_datetime(data_copy[col], errors="coerce")
data_copy.info()
````

````python
#displaying no of null values for each column with datatype datetime
for col in date_columns:
  print(f"No of null values in {col} is {data_copy[col].isna().sum()}")
````

````python
#dropping column pq_first_sent_to_client_date since more than 25 percent null values are present
#and it is not that important for this problem statement
data_copy.drop(columns=["pq_first_sent_to_client_date"], inplace=True)

#dropping column po_sent_to_vendor_date since more than 50 percent null values are present
data_copy.drop(columns=["po_sent_to_vendor_date"], inplace=True)
````

````python
data_copy.head()
````

````python
#removing whitespaces from values of columns that have strings in them
for col in data_copy.select_dtypes(include="object").columns:
  data_copy[col] = data_copy[col].str.strip()
````

````python
#creating a new column delivery_delay_days
data_copy["delivery_delay_days"]=(data_copy["delivered_to_client_date"]-data_copy["scheduled_delivery_date"]).dt.days

#creating a column delivery_performance with values "Delayed" or "on-time"
data_copy["delivery_performance"] = np.where(data_copy["delivery_delay_days"] > 0, "Delayed", "On-time")
````

````python
#sorting the values in ascending order based on asndn number then by order id
data_copy.sort_values(by=['asndn_','id'],ascending=True, inplace = True, ignore_index = True)

#removing data with same asndn number because total weight and freight cost are recorded only once for same asn/dn number
weights_and_freights = data_copy.drop_duplicates(subset=['asndn_'], keep='first')



````

````python
#converting data type of weight_kilograms from object to numerical
weights_and_freights["weight_kilograms"] = pd.to_numeric(
    weights_and_freights["weight_kilograms"],
    errors="coerce"
)

#converting data type of freight_cost_usd from object to numerical
weights_and_freights["freight_cost_usd"] = pd.to_numeric(
    weights_and_freights["freight_cost_usd"],
    errors="coerce"
)

#filling null values in weight_kilograms with median
weights_and_freights["weight_kilograms"].fillna(weights_and_freights["weight_kilograms"].median(), inplace=True)

#filling null values in freight_cost_usd with median
weights_and_freights["freight_cost_usd"].fillna(weights_and_freights["freight_cost_usd"].median(), inplace=True)

weights_and_freights[['weight_kilograms', "freight_cost_usd"]].head()
````

````python
#dropping columns wieght_kilograms and freight_cost_usd
#so that when merging with weights_and_freights these columns are not repeated
data_copy=data_copy.drop(columns=["weight_kilograms", "freight_cost_usd"])

#removing all columns from weights_and_freights except these three
#so that when merging with data_copy these columns are not repeated
weights_and_freights=weights_and_freights[["asndn_", "weight_kilograms", "freight_cost_usd"]]

#merging data_copy and weights_and_freights on asndn_
data_copy=pd.merge(data_copy,weights_and_freights, on="asndn_", how="inner")
data_copy.shape
````

````python
#checking for outliers in weight_kilograms using boxplot
fig, axes=plt.subplots(1,2)
axes[0].boxplot(data_copy["weight_kilograms"], orientation="horizontal")
axes[0].set_title('Weights(kg)')

#checking for outliers in freight_cost_usd using boxplot
axes[1].boxplot(data_copy["freight_cost_usd"], orientation="horizontal")
axes[1].set_title('Freight Cost(USD)')

plt.tight_layout()
plt.show()
````

#### Chart - 12 Frequency Distribution of Delivery Performance for Various FulFillment Methods

````python
#code to make 4 subplots
fig = make_subplots(
    rows=2,
    cols=2,
    subplot_titles=("Bar Chart", "Direct Drop Delivery Performance", "", "From RDC Delivery Performance"), # Updated titles for clarity
    specs=[[{'type':'xy'}, {'type':'domain'}],
           [{'type':'xy'}, {'type':'domain'}]] # Specifying 'domain' for pie chart positions
)

#defining a variable to use to color code the bars and pie charts
color_map = {
    "On-time": "#2ECC71",
    "Delayed": "#E74C3C"
}
#stores the count of each group after grouping by fulfill via and delivery performance
counts = (
    data_copy
    .groupby(["fulfill_via", "delivery_performance"])
    .size()
    .reset_index(name="count")
)

#for loop is used to create one bar per category
for perf in counts["delivery_performance"].unique():#extracts unique values in delivery performance
    subset = counts[counts["delivery_performance"] == perf] #filters data to one unique value of delivery performance

    #plotting a bar graph to show frequency distribution of delivery performance for fulfillments methods at 1,1
    fig.add_trace(
        go.Bar(
            x=subset["fulfill_via"],
            y=subset["count"],
            name=perf,
            showlegend=True,
            marker=dict(color=[
                color_map[l] for l in subset["delivery_performance"]])
        ),
        row=1, col=1
    )

#stores data that are fulfilled via direct drop
d_drop_delivery=data_copy[data_copy["fulfill_via"]=="Direct Drop"]

#stores count of each unique value of delivery performance for direct deliver
counts_direct_drop = (
    d_drop_delivery["delivery_performance"]
    .value_counts()
    .reset_index()
)

counts_direct_drop.columns = ["delivery_performance", "count"]

#plotting a pie chart at 1,2 that shows delivery performance of direct drop
fig.add_trace(
    go.Pie(
        labels=counts_direct_drop["delivery_performance"],
        values=counts_direct_drop["count"],
        hole=0.4,
        hovertemplate="%{label}: %{value} (%{percent})",
        name="Direct Drop",
        marker=dict(colors=[
            color_map[l] for l in counts_direct_drop["delivery_performance"]])
    ),
    row=1, col=2
  )

#stores data that are fulfilled via rdc
rdc_delivery=data_copy[data_copy["fulfill_via"]=="From RDC"]

#stores count of each unique value of delivery performance for rdc
counts_rdc = (
    rdc_delivery["delivery_performance"]
    .value_counts()
    .reset_index()
)

counts_rdc.columns = ["delivery_performance", "count"]

#plotting a pie chart at 2,2 that shows delivery performance of rdc
fig.add_trace(
    go.Pie(
        labels=counts_rdc["delivery_performance"],
        values=counts_rdc["count"],
        hole=0.4,
        hovertemplate="%{label}: %{value} (%{percent})",
        name="From RDC",
        marker=dict(colors=[
            color_map[l] for l in counts_rdc["delivery_performance"]])
    ),
    row=2, col=2
  )

#updates layout of all the subplots
fig.update_layout(
    barmode="group",
    title="Fulfill Via vs Delivery Performance",
    template="simple_white",
    title_x=0.5,
    height=700,
    showlegend=True
)

# Update x-axis and y-axis labels for the bar chart
fig.update_xaxes(title_text="Fulfill Via", row=1, col=1)
fig.update_yaxes(title_text="Count", row=1, col=1)

fig.show()
````
